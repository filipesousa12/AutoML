```python
import pandas as pd
import numpy as np
import h2o
```

```python
df=pd.read_excel('df_clean.xlsx')
```

```python
df['Transported'] = df['Transported'].astype('category')
```

```python
df.describe()
```

```python
h2o.init()
```

```python
from h2o.automl import H2OAutoML
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test= train_test_split(df,test_size = 0.25, random_state = 0)
```

```python
df_train_h2o = h2o.H2OFrame(df_train)
df_test_h20 = h2o.H2OFrame(df_test)
df_h2o= h2o.H2OFrame(df)
```

```python
df_train_h2o["Transported"]=df_train_h2o["Transported"].asfactor()
df_test_h20["Transported"]=df_test_h20["Transported"].asfactor()
```

```python
df_train_h2o.types
```

```python
y="Transported"
x=df_h2o.columns
x.remove(y)
```

```python
aml=H2OAutoML(max_models=20,seed=10,exclude_algos=["DeepLearning"],verbosity='info',nfolds=-1)
```

```python
aml.train(x=x,y=y,training_frame=df_train_h2o,validation_frame=df_test_h20)
```

```python
lb=aml.leaderboard
```
